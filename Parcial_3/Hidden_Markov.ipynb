```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations
from itertools import combinations
from itertools import combinations_with_replacement


States = np.array([0,1]) # 0(J), 1(B)
Prior = np.array([0.2,0.8])

T = np.array([[0.8,0.2],[0.2,0.8]])
E = np.array([[0.5,0.9],[0.5,0.1]])

Obs = np.array([1,0,0,0,1,0,1,0])

DictH = {0:'Justa',1:'Sesgada'}
print(DictH[0])
DictO = {0:'Cara',1:'Sello'}
print(DictO[0])

def GetStates(State,N):
    CStates = list(combinations_with_replacement(State,N))
    Permu = []
    for it in CStates:
      p = list(permutations(it,N))

      for i in p:
        if i not in Permu:
          Permu.append(i)
    CStates = np.array(Permu)
    return CStates

HiddenStates = GetStates(States,len(Obs))
print(HiddenStates)

def GetProb(T,E,Obs,State,Prior):

  n = len(Obs)
  p = 1.
  p *= Prior[ State[0] ]

  for i in range(n-1):
    p *= T[ State[i+1], State[i] ]
  for i in range(n):
    p *= E[ Obs[i], State[i] ]
  return p

P = np.zeros(HiddenStates.shape[0])

for i in range(P.shape[0]):
  P[i] = GetProb(T,E,Obs,HiddenStates[i],Prior)

print(P)
ii = np.where( P == np.max(P) )
print(HiddenStates[ii],P[ii])
max_value = np.max(P)

#c)

plt.axhline(y=max_value, color='teal', linestyle='--', label='Máximo')
plt.legend()
plt.plot(P, color="plum")
plt.legend()
plt.grid(True)
plt.xlabel('Estados ocultos')
plt.ylabel('Probabilidad')
plt.title('Probabilidad de cada estado oculto')
plt.show()
pobs = np.sum(P)
print(pobs)
ObsStates = GetStates([0,1],8)
print(ObsStates)

Nobs = ObsStates.shape[0]
PObs = np.zeros(Nobs)

for j in range(Nobs):

  dim = HiddenStates.shape[0]
  P = np.zeros(dim)

  for i in range(dim):
    P[i] = GetProb(T,E,ObsStates[j],HiddenStates[i],Prior)

  PObs[j] = np.sum(P)
print(PObs)
plt.plot(PObs)


plt.show()
ii = np.where( PObs == np.max(PObs) )
print(ObsStates[ii],PObs[ii])

print(np.sum(PObs))
```

Si tuviera que estimar las probabilidades de transición y emisión utilizando técnicas de aprendizaje automático, ¿qué enfoque metodológico basado en datos propondría? Si desea ilustrar su respuesta con datos sintéticos, también es valido.

```python
import numpy as np
import random

# Función para generar secuencias sintéticas
def generate_sequence(T, E, states, observations, length):
    n_states = len(states)
    current_state = np.random.choice(range(n_states))  # Elegir un estado inicial aleatorio
    state_sequence = [current_state]
    obs_sequence = []

    for _ in range(length):
        # Generar observación basada en el estado actual
        obs = np.random.choice(len(observations), p=E[current_state])
        obs_sequence.append(obs)

        # Cambiar al siguiente estado basado en la matriz de transición
        current_state = np.random.choice(range(n_states), p=T[current_state])
        state_sequence.append(current_state)

    readable_states = [states[s] for s in state_sequence]
    readable_obs = [observations[o] for o in obs_sequence]
    return readable_states, readable_obs

# Clase GeneticHMM
class GeneticHMM:
    def __init__(self, states, observations, seq_obs, pop_size=50, generations=100, mutation_rate=0.1):
        self.states = states
        self.observations = observations
        self.seq_obs = seq_obs
        self.n_states = len(states)
        self.n_obs = len(observations)
        self.pop_size = pop_size
        self.generations = generations
        self.mutation_rate = mutation_rate
        self.population = self.initialize_population()

    def initialize_population(self):
        population = []
        for _ in range(self.pop_size):
            T = self.random_normalized_matrix(self.n_states, self.n_states)
            E = self.random_normalized_matrix(self.n_states, self.n_obs)
            population.append((T, E))
        return population

    def random_normalized_matrix(self, rows, cols):
        matrix = np.random.rand(rows, cols)
        return matrix / matrix.sum(axis=1, keepdims=True)

    def fitness(self, individual):
        T, E = individual
        n = len(self.seq_obs)
        alpha = np.zeros((n, self.n_states))  # Algoritmo forward

        # Inicialización
        alpha[0, :] = E[:, self.seq_obs[0]]
        for t in range(1, n):
            alpha[t, :] = E[:, self.seq_obs[t]] * np.dot(alpha[t - 1, :], T)


        # En lugar de solo usar log-sum-exp, puedes sumar el log de las probabilidades individuales
        log_prob = np.sum(np.log(np.sum(np.exp(alpha), axis=1)))
        return log_prob

        # Usar log-sum-exp para evitar problemas numéricos
        log_sum_exp = np.log(np.sum(np.exp(alpha[-1, :])))
        return log_sum_exp
    def select(self):
        fitness_values = np.array([self.fitness(ind) for ind in self.population])
        # Realizamos una selección por torneo, donde se seleccionan dos individuos y el mejor es elegido
        tournament_size = min(5, len(self.population))  # Change this line to ensure tournament_size is within bounds
        selected_parents = []
        for _ in range(self.pop_size):
            tournament = random.sample(range(len(self.population)), tournament_size) # Change this line
            best_individual = max(tournament, key=lambda idx: fitness_values[idx])
            selected_parents.append(self.population[best_individual])
        return selected_parents

    def crossover(self, parent1, parent2):
        T1, E1 = parent1
        T2, E2 = parent2
        n_states, n_obs = T1.shape

        # Cruce de matrices de transición y emisión
        crossover_point = random.randint(1, n_states - 1)  # Punto de cruce aleatorio
        T_child = np.vstack((T1[:crossover_point], T2[crossover_point:]))
        E_child = np.vstack((E1[:crossover_point], E2[crossover_point:]))
        return T_child, E_child

    def mutate(self, individual):
        T, E = individual
        if random.random() < self.mutation_rate:
            T += np.random.normal(0, 0.3, T.shape)  # Incremento en la varianza
            T = np.abs(T)
            T /= T.sum(axis=1, keepdims=True)
        if random.random() < self.mutation_rate:
            E += np.random.normal(0, 0.3, E.shape)  # Incremento en la varianza
            E = np.abs(E)
            E /= E.sum(axis=1, keepdims=True)
        return T, E

    def evolve(self):
        for gen in range(self.generations):
            selected_parents = self.select()
            new_population = []
            for i in range(0, len(selected_parents), 2):
                parent1, parent2 = selected_parents[i], selected_parents[i + 1]
                child = self.crossover(parent1, parent2)
                child = self.mutate(child)
                new_population.append(child)
            self.population = new_population

            best_individual = max(self.population, key=self.fitness)
            print(f"Generación {gen + 1}, Mejor Fitness: {self.fitness(best_individual):.4f}")

        best_T, best_E = max(self.population, key=self.fitness)
        return best_T, best_E

# Parámetros del problema
states = ["Justa", "Sesgada"]
observations = ["Cara", "Sello"]
T_real = np.array([[0.8, 0.2], [0.2, 0.8]])
E_real = np.array([[0.5, 0.5], [0.9, 0.1]])
np.random.seed(42)
synthetic_states, synthetic_obs = generate_sequence(T_real, E_real, states, observations, 100)

# Crear y ejecutar el algoritmo genético
ga_hmm = GeneticHMM(states, observations, [observations.index(o) for o in synthetic_obs])
best_T, best_E = ga_hmm.evolve()

print("Matriz de Transición Estimada (T):\n", best_T)
print("Matriz de Emisión Estimada (E):\n", best_E)
```
